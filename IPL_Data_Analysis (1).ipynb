```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# -------------------------------
# CSV DATA (Sample IPL Dataset)
# -------------------------------
data = """season,team1,team2,toss_winner,toss_decision,winner,player_of_match
2020,MI,CSK,MI,bat,MI,Rohit Sharma
2020,RCB,SRH,SRH,field,SRH,David Warner
2020,CSK,RR,RR,bat,RR,Steve Smith
2021,MI,RCB,RCB,field,RCB,AB de Villiers
2021,CSK,SRH,CSK,bat,CSK,Suresh Raina
2021,RR,MI,MI,field,MI,Hardik Pandya
2022,GT,RR,RR,bat,GT,Shubman Gill
2022,LSG,CSK,CSK,field,LSG,KL Rahul
2022,RCB,MI,RCB,bat,RCB,Virat Kohli
2023,CSK,GT,GT,field,CSK,Ruturaj Gaikwad
2023,MI,RR,MI,bat,MI,Suryakumar Yadav
2023,SRH,RCB,SRH,field,RCB,Faf du Plessis
"""

# Convert to DataFrame
df = pd.read_csv(StringIO(data))

print("Dataset Preview:")
print(df.head())

# -------------------------------
# 1. Top Winning Teams
# -------------------------------
top_teams = df['winner'].value_counts()

plt.figure()
top_teams.plot(kind='bar')
plt.title("Top Winning Teams")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 2. Toss Decision Analysis
# -------------------------------
plt.figure()
sns.countplot(x='toss_decision', data=df)
plt.title("Toss Decision Count")
plt.show()

# -------------------------------
# 3. Top Players of Match
# -------------------------------
top_players = df['player_of_match'].value_counts()

plt.figure()
top_players.plot(kind='bar')
plt.title("Top Players of the Match")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 4. Matches Per Season
# -------------------------------
matches_per_season = df['season'].value_counts().sort_index()

plt.figure()
matches_per_season.plot(kind='line', marker='o')
plt.title("Matches Per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()

# -------------------------------
# 5. Toss Winner vs Match Winner
# -------------------------------
toss_match = df[df['toss_winner'] == df['winner']]

print("\nToss Winner also Won Match:", len(toss_match))
print("Total Matches:", len(df))
print("Win Percentage:", (len(toss_match)/len(df))*100)
```
